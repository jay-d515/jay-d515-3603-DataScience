Weâ€™ll use the famous **Penguins dataset** (thanks to Seaborn).

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
penguins = pd.read_csv("penguins.csv")
penguins.head()
````

## 1. Explore Distributions

First, letâ€™s look at the distribution of **body mass**.

````python
sns.histplot(data=penguins, x="body_mass_g", kde=True, hue="species")
plt.title("Distribution of Body Mass by Species")
plt.show()
````

ðŸ‘‰ **Task:** Do some species have higher average body mass than others?

Adelle's tend to have a higher average body mass than Cinstraps and Gentoos.

## 2. Relationships Between Variables

Scatterplots help us see trends between two quantitative variables.

````python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Flipper Length vs Body Mass")
plt.show()
````

ðŸ‘‰ **Task:** Which species tends to have longer flippers AND higher body mass?

Gentoos tend to have longer flippers and higher body mass.

## 3. Box Plots for Outliers

Box plots are great for spotting outliers.

````python
sns.boxplot(data=penguins, x="species", y="bill_length_mm", palette="Set2")
plt.title("Bill Length by Species")
plt.show()
````

ðŸ‘‰ **Task:** Can you spot any outliers in the bill length data?

There is an outlier in the box plot for Gentoos.

## 4. Correlation Heatmap

Correlation heatmaps let us see overall variable relationships.

````python
corr = penguins.dropna().corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
````

ðŸ‘‰ **Task:** Which variables appear to be strongly correlated?

Body mass and flipper length appear to be strongly correlated.

````python
pokemon = pd.read_csv("pokemon.csv")
pokemon.head()
````

````python
# Create 2 plots that reveal a trend

# This boxplot illustartes that Type 1 Dragon types typically have higher total stats
sns.boxplot(data=pokemon, x='Type 1', y='Total')
plt.xticks(rotation=90)
plt.title('Total Stats by PokÃ©mon Type')
plt.show()
````

````python
# This illustrates that Dragon and Psychic types tended to have a higher number of Legendary PokÃ©mon
legend_count = pokemon[pokemon["Legendary"]].groupby("Type 1").size().sort_values(ascending=False)
legend_count.plot(kind='bar', figsize=(10,5))
plt.title("Number of Legendary PokÃ©mon per Type")
plt.ylabel("Count")
plt.show()
````

````python
# Create 1 plot that highlights possible outliers.
stats_cols = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
plt.figure(figsize=(12,6))
sns.boxplot(data=pokemon[stats_cols])
plt.title("Outliers Across PokÃ©mon Stat Categories")
plt.show()
````

The boxplot, 'Total Stats by PokÃ©mon Type', revealed that Type 1 Dragon type pokÃ©mon tend to have higher total stats compared to other Type 1 pokÃ©mon. The barplot, 'Number of Legendary PokÃ©mon per Type', revealed that Dragon and Psychic type pokÃ©mon tend to have a higher number of Legendary PokÃ©mon. Finally, the boxplot, 'Ouliers across PokÃ©mon Stat Categories', across all the stat categories, HP had the most outliers leaning towards both the lower and higher ends.
