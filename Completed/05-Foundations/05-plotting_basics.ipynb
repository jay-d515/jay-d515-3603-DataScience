## 1. Setup

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

## 2. Line Plot

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
````

âœ… **Your Turn**: Plot `cos(x)` on the same graph with a different color and label.

````python
x1 = np.linspace(0, 10, 100)
y1 = np.cos(x1)

plt.plot(x1, y1, label="cos(x)", color="red")
plt.title("My Line Plot")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.legend()
plt.show()
````

## 3. Scatter Plot

````python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
````

âœ… **Your Turn**: Generate two sets of random points and plot them in different colors on the same scatter plot.

````python
# generate two sets of random points in different colors
x1 = np.random.rand(50)
y1 = np.random.rand(50)
sizes1 = 300 * np.random.rand(50)
colors1 = np.random.rand(50)

x2 = np.random.rand(50)
y2 = np.random.rand(50)
sizes2 = 300 * np.random.rand(50)
colors2 = np.random.rand(50)

# plots teh random points on the same scatter plot
plt.scatter(x1, y1, s=sizes1, c=colors1, alpha=0.6, cmap='viridis')
plt.scatter(x2, y2, s=sizes2, c=colors2, alpha=0.6, cmap='viridis')
plt.title("My Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
````

## 4. Bar Plot

````python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
````

## 5. Customizing Plots

````python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
````

âœ… **Your Turn**: Customize your own plot with labels, legend, and grid lines.

````python
x1 = np.linspace(0, 2*np.pi, 100)
my_y1 = np.sin(x)
my_y2 = np.tan(x)

plt.plot(x1, my_y1, label="sin(x)", linestyle="--", color="green")
plt.plot(x1, my_y2, label="tan(x)", linestyle=":", color="orange")
plt.title("My Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
````
